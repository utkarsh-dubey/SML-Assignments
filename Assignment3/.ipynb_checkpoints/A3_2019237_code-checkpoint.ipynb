```python
# Visualize 5 samples from each class in the form of images.
def display_image(X, Y, to_display):
    images = X.reshape(len(X), 3, 32, 32).transpose(0, 2, 3, 1)
    for i in range(0, 50):
        plt.imshow(images[to_display[i]])
        plt.title("{}".format(labels[Y[to_display[i]]]))
        plt.show()

to_display = []

for i in range(10):
    index = np.where(i == Y_train)[0]
    to_display += np.random.choice(index, size=5, replace=False).tolist()

display_image(X_train, Y_train, to_display)
```

```python
def classwise_accuracy(Y_test, predictions, labels):
    cm = confusion_matrix(Y_test, predictions)
    accuracies = cm.diagonal()/cm.sum(axis=1)
    for i in range(10):
        print("Accuracy for " + str(labels[i]) + " =", accuracies[i])
```

```python
# Imports
from mlxtend.data import loadlocal_mnist
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.preprocessing import StandardScaler
```

```python
train_images,train_labels = loadlocal_mnist(images_path='./mnist/train-images-idx3-ubyte', labels_path='./mnist/train-labels-idx1-ubyte')
test_images,test_labels = loadlocal_mnist(images_path='./mnist/t10k-images-idx3-ubyte', labels_path='./mnist/t10k-labels-idx1-ubyte')
```

```python
accuracies = []
components = [15, 8, 3]
pcas = []
clfs = []
for i in range(len(components)):
    scaler = StandardScaler()
    scaler.fit_transform(train_images)
    pca = PCA(n_components = components[i])
    transform = pca.fit_transform(train_images)
    lda = LinearDiscriminantAnalysis()
    lda.fit(transform, train_labels)
    test_transform = pca.transform(test_images)
    y = lda.predict(test_transform)
    accuracies.append(100*np.sum(y == test_labels)/len(test_labels))

plt.plot(components, accuracies)
plt.xlabel("N components")
plt.ylabel("Accuracy")
plt.title("Accuracy vs PCA Components")
```

```python
# Q3
from mlxtend.data import loadlocal_mnist
from numpy import linalg
train_images,train_labels = loadlocal_mnist(images_path='./fminst/train-images-idx3-ubyte', labels_path='./fminst/train-labels-idx1-ubyte')
test_images,test_labels = loadlocal_mnist(images_path='./fminst/t10k-images-idx3-ubyte', labels_path='./fminst/t10k-labels-idx1-ubyte')
```

```python
class FDA:
    W = None
    eigen_values = None
    Sw, Sb = None, None

    def __init__(self):
        W = None
        
    def fit(self, X, Y):
        Sw = self.get_scatter_within(X, Y)
        Sb = self.get_scatter_between(X, Sw)
        
        self.Sw = Sw
        self.Sb = Sb
        
        # computing w = eigen vectors of Sw_inverse * Sb
        Sw_inv_Sb = np.matmul(np.linalg.inv(Sw), Sb)
        eig_values, eig_vectors = linalg.eigh(Sw_inv_Sb)

        # correcting the order
        idx = eig_values.argsort()[::-1]   
        eig_vectors = eig_vectors[:,idx]
        eig_values = eig_values[idx]

        # considering top rank vectors from the eigen vectors
        rank = linalg.matrix_rank(Sb)
        eig_vectors = eig_vectors[:,:rank]
        eig_values  = eig_values[:rank]

        self.eigen_values = eig_values
        self.W = eig_vectors

    def get_scatter_within(self, X, Y):
        Si = []
        unique_Y = np.unique(Y)
        for label in unique_Y:
            indexes = np.where(Y == label)[0]
            selected_X = X[indexes]
            selected_X_trans = selected_X.T
            scatter_matrix = np.cov(selected_X_trans, ddof=0) * selected_X_trans.shape[1]
            Si.append(scatter_matrix)
        Sw = np.zeros(Si[0].shape)
        for scatter_matrix in Si:
            Sw += scatter_matrix
        return Sw
    
    def get_scatter_between(self, X, Sw):
        X_tran = X.T
        St = np.cov(X_tran, ddof=0) * X_tran.shape[1]
        Sb = St - Sw
        return Sb
```

```python
myFDA = FDA()
myFDA.fit(train_images, train_labels)

W = myFDA.W

X_train_projected = np.matmul(W.T, train_images.T).T
X_test_projected  = np.matmul(W.T, test_images.T).T

clf = LinearDiscriminantAnalysis()
clf.fit(X_train_projected, train_labels)

print("Accuracy =", clf.score(X_test_projected, test_labels))
```
