```python
import numpy as np
from math import *
import scipy.stats
import matplotlib.pyplot as plt
```

```python
u1=[0.5,0.8]
u2=[0.9,0.2]
class1 = np.array([scipy.stats.bernoulli.rvs(u1[0], size = 100), scipy.stats.bernoulli.rvs(u1[1], size = 100)])
class2 = np.array([scipy.stats.bernoulli.rvs(u2[0], size = 100), scipy.stats.bernoulli.rvs(u2[1], size = 100)])
print("class1\n")
print(class1)
print(class1.shape)
print("class2\n")
print(class2)
print(class2.shape)
```

```python
def MLE(data):
    d,n=data.shape
    theta=[0]*d
    for i in range(d):
        theta[i]=np.sum(data[i])/n
    return theta
```

```python
trainData = class1[:,:50]
testData = class1[:,50:]
print(trainData.shape)
print(testData.shape)
```

```python
mle1 = np.array(MLE(trainData))
print(mle1,"\n")
print(mle1.shape)
```

```python
#for n samples at a time
mles = [[],[]]
values = [[],[]]
for i in range(50):
    values[0].append(trainData[0][i])
    values[1].append(trainData[1][i])
    temp = MLE(np.array(values))
    mles[0].append(temp[0])
    mles[1].append(temp[1])

plt.plot(range(1,51),mles[0])
plt.title("Theta 1 vs N")
plt.show()

plt.plot(range(1,51),mles[1])
plt.title("Theta 2 vs N")
plt.show()
```

```python
trainData2 = class2[:,:50]
testData2 = class2[:,50:]
print(trainData2.shape)
print(testData2.shape)

mle2 = np.array(MLE(trainData2))
print(mle2,"\n")
print(mle2.shape)

#for n samples at a time
mles2 = [[],[]]
values2 = [[],[]]
for i in range(50):
    values2[0].append(trainData2[0][i])
    values2[1].append(trainData2[1][i])
    temp = MLE(np.array(values2))
    mles2[0].append(temp[0])
    mles2[1].append(temp[1])

plt.plot(range(1,51),mles2[0])
plt.title("Theta 1 vs N")
plt.show()

plt.plot(range(1,51),mles2[1])
plt.title("Theta 2 vs N")
plt.show()
```

```python
def disc(one, two, x1, x2):
    return (one ** x1) * ((1 - one) ** (1 - x1)) * (two ** x2) * ((1 - two) ** (1 - x2))
```

```python
predictions = 0
for i in range(50):
    one = log(disc(sum(mles[0])/50,sum(mles[1])/50,testData[0][i],testData[1][i]))
    two = log(disc(sum(mles2[0])/50,sum(mles2[1])/50,testData[0][i],testData[1][i]))
    
    if((one-two)>0):
        predictions+=1
        
print("Accuracy",predictions*2)
    
```
