```python
import numpy as np
from math import *
import matplotlib.pyplot as plt
```

```python
x=np.matrix([[1,3],[4,6]])
print(x)
```

```python
xMean = np.mean(x,axis=1)
print(xMean)
```

```python
Xc = np.subtract(x,xMean)
print(Xc)
```

```python
covXc = (Xc*Xc.transpose())/Xc.shape[0]
print(covXc)
```

```python
w,U = np.linalg.eig(covXc)
print(w)
print(U)
```

```python
Y = U.transpose()*Xc
print(Y)
```

```python
#part b
temp = U*Y+xMean
print(temp)
```

```python
MSE = ((np.array(x)-np.array(temp))**2).mean(axis=None)
print("MSE\n",MSE)
```

```python
#part d
inputs = list(map(float,input().split()))
d = int(inputs[0])
n = int(inputs[1])
mu = inputs[2]
sigma = inputs[3]

x = []

for i in range(d):
    x.append(np.random.normal(mu, sigma, n))

x = np.matrix(x)
print(x.shape)
print("\n",x)
```

```python
#part e
xMean = np.mean(x,axis=1)
print(xMean)
```

```python
Xc = np.subtract(x,xMean)
print(Xc)
```

```python
covXc = (Xc*Xc.transpose())/Xc.shape[0]
print(covXc)
```

```python
w,U = np.linalg.eig(covXc)
print(w)
print("eigen vectors",U)
```

```python
Y = U.transpose()*Xc
print("Y\n",Y)
```

```python
#part b
temp = U*Y+xMean
print(temp)
MSE = ((np.array(x)-np.array(temp))**2).mean()
print("MSE\n",MSE)
```

```python
xplot=[]
yplot=[]
w,U = np.linalg.eig(covXc)
print(U.shape)
# newU = np.transpose(U)
print(U)
for i in range(1,d+1):
    Y = U[:i]*Xc
    temp = U[:i].transpose()*Y+xMean
    yplot.append(((np.array(x)-np.array(temp))**2).mean())
    xplot.append(len(yplot))

plt.plot(xplot,yplot)
plt.title("MSE vs no of eigenvectors")
plt.show()
```
